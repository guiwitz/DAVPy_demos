# So easy, *voilà*!

In this example notebook, we demonstrate how Voilà can render custom interactive matplotlib figures using the [ipympl](https://github.com/matplotlib/ipympl) widget.

```python
import pandas as pd
```

```python
penguins = pd.read_csv('https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv')
```

```python
%matplotlib widget
import ipympl

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive
from IPython.display import display
```

```python
penguins.head(3)
```

```python
sns.scatterplot(penguins, x='bill_length_mm', y='bill_depth_mm');
```

```python
def plot_func(frequency=1.0):
    #plt.figure(fig)
    plt.clf()
    t = np.linspace(0, 1, 1000)
    plt.plot(t, np.sin(2 * np.pi * frequency * t))
    plt.ylim(-1.5, 1.5)
    plt.title('Interactive Plot')
    #plt.show()
```

```python
fig, ax = plt.subplots()
```

```python
interactive_plot = interactive(plot_func, frequency=(0.1, 10.0))
display(interactive_plot)
```
